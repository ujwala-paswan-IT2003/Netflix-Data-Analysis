```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df=pd.read_csv('netflixdailytop_dataset.csv')
df=df.head(10)
sns.scatterplot(x='Rank',y='Days In Top 10',data=df,palette="CMRmap",ci=None)
plt.xlabel("Rank")
plt.ylabel("Days In Top 10")
plt.title("Scatter Plot(6207 UJWALA)")
plt.show()
```
